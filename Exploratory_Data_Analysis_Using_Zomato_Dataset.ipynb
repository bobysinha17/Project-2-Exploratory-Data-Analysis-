```python
df_country = pd.read_excel('Country-Code.xlsx') # This is the second dataset titled as 'Country-Code'. Therefore, I am reading that particular dataset.
```

```python
df_country.head(10) #Top 10 entries.
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country, on= 'Country Code', how ='left') #Since we are having two different datasets with the same feature name as 'Country Code', therefore we are trying to merge in one single dataframe called final_df.
```

```python
final_df.head(5)
```

```python
#To check the datatype info.
final_df.dtypes
```

```python
#Columns/feature of the dataset.
final_df.columns 
```

```python
final_df.Country.value_counts() #In country column how many country are there and its value
```

```python
country_name= final_df.Country.value_counts().index
```

```python
country_value= final_df.Country.value_counts().values
```

```python
#Plot Pie Chart:
plt.pie(country_value,labels= country_name)
```

```python
#Plot Pie Chart: I want to see Top 3 countries 
plt.pie(country_value[:3],labels= country_name[:3],autopct= '%1.2f%%')
```

Interpretation: The most orders are from India followed bt United States and lastly United Kingdom

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns= {0:'Rating Count'})
```

```python
ratings
```

```python
# Observation:
1. When rating is between from 4.5 to 4.9 is Excellent
2. When rating is between from 4.0 to 4.4 is Very Good
3. When rating is between 3.5 to 3.9 us Good
4. o rating has been given by many people
```

```python
ratings.head()
```

```python
import matplotlib
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
matplotlib.rcParams['figure.figsize'] = (10,10) #To set the figure size
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings,hue ='Rating color', palette=palette=['blue','red','orange','yellow','green','green'])
```

```python
## Count Plot Y should be categorical variable:
sns.countplot(x='Rating color', data= ratings, palette=['blue','red','orange','yellow','green','green'])
```
